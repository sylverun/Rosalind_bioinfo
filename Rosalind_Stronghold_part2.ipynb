```python
def open_fasta(path, mode):
    fasta_file = open(path, mode)
    sequences = {}

    for line in fasta_file:
        line = line.strip()  # Remove whitespace at the beginning and end of the line
        if line.startswith(">"):  # Check if the line is a description line
            current_sequence_name = line[1:]  # Remove the ">" symbol
            sequences[current_sequence_name] = ""  # Add the sequence name to the dictionary
        else:
            sequences[current_sequence_name] += line  # Add the sequence data to the current sequence name key in the dictionary

    fasta_file.close()
    return sequences
```

```python
def find_longest_common_substring(dico):
    shortest_string = min(dico.values(), key=lambda x: len(x))
    n = len(shortest_string)
    longest_substring = ""
    for i in range(n):
        for j in range(i + 1, n + 1):
            substring = shortest_string[i:j]
            if len(substring) > len(longest_substring) and all(substring in s for s in list(dico.values())):
                longest_substring = substring
    return longest_substring
```

### Mendel 2nd Problem

```python
### they are a 25% probability to obtain an Aa Bb offspring so we need only a binomial law to compute the probability at the k_th generation
from scipy.special import binom

def binomial_offspring(k, N):
    def p(n, k):
        return binom(2**k, n) * 0.25**n * 0.75**(2**k - n) # 2k because 2 offsprings by reproduction
    return 1 - sum(p(n, k) for n in range(N))
```

```python
# Getting the RNA_string for both strands

def trans_into_RNA(strand):
    RNA=''
    for char in (strand):
        if char == 'T':
            RNA += 'U'
        else:
            RNA += char
    return RNA
RNA_string_N= trans_into_RNA(string)
RNA_string_C= trans_into_RNA(complementary_strand)
print(RNA_string_N)
print(RNA_string_C)
```

```python
# Identification of the chunks that will be convert to protein

def Chunks_to_traduct(RNA):
    pattern = re.compile(r'(?=(AUG(?:...)*?)(?=UAA|UAG|UGA))')
    # Search for matches in the protein sequence
    matches = pattern.findall(RNA)
    return matches

trad_RNA_N= Chunks_to_traduct(RNA_string_N)
trad_RNA_C= Chunks_to_traduct(RNA_string_C)
print(trad_RNA_N)
print(trad_RNA_C)
```

```python
def Traduction_to_protein(RNA_chunks):
    dico_RNA_to_prot={'UUU' :'F','CUU' : 'L', 'AUU' : 'I', 'GUU' : 'V',
    'UUC' : 'F', 'CUC' : 'L', 'AUC' : 'I', 'GUC' : 'V',
    'UUA' : 'L', 'CUA' : 'L', 'AUA' : 'I', 'GUA' : 'V',
    'UUG' : 'L', 'CUG' : 'L', 'AUG' : 'M', 'GUG' : 'V',
    'UCU' : 'S', 'CCU' : 'P', 'ACU' : 'T', 'GCU' : 'A',
    'UCC' : 'S', 'CCC' : 'P', 'ACC' : 'T', 'GCC' : 'A',
    'UCA' : 'S', 'CCA' : 'P', 'ACA' : 'T', 'GCA' : 'A',
    'UCG' : 'S', 'CCG' : 'P', 'ACG' : 'T', 'GCG' : 'A',
    'UAU' : 'Y', 'CAU' : 'H', 'AAU' : 'N', 'GAU' : 'D',
    'UAC' : 'Y', 'CAC' : 'H', 'AAC' : 'N', 'GAC' : 'D',
    'UAA' : ' ', 'CAA' : 'Q', 'AAA' : 'K', 'GAA' : 'E',
    'UAG' : ' ', 'CAG' : 'Q', 'AAG' : 'K', 'GAG' : 'E',
    'UGU' : 'C', 'CGU' : 'R', 'AGU' : 'S', 'GGU' : 'G',
    'UGC' : 'C', 'CGC' : 'R', 'AGC' : 'S', 'GGC' : 'G',
    'UGA' : ' ', 'CGA' : 'R', 'AGA' : 'R', 'GGA' : 'G',
    'UGG' : 'W', 'CGG' : 'R', 'AGG' : 'R', 'GGG' : 'G',}
    n = 3 
    list_proteins=[]
    for chunk in RNA_chunks:
        prot_seq=''
        codons = [chunk[i:i+n] for i in range(0, len(chunk),n)]
        for codon in codons:
            prot_seq += dico_RNA_to_prot[codon]
        list_proteins.append(prot_seq)
    return list_proteins


List_prot_N= Traduction_to_protein(trad_RNA_N)
List_prot_C= Traduction_to_protein(trad_RNA_C)

print(List_prot_N)
print(List_prot_C)
```

### Locating Restriction Sites Problem

```python
seqs_3= open_fasta(r"C:\Users\sylv_\Desktop\Rosalind download\rosalind_revp.txt", "r")

# Getting the  strand
string=list(seqs_3.values())[0]
dico_reverse={'A':'T',
             'G': 'C',
             'C': 'G',
             'T': 'A'}

liste_palindrome=[]
for idx in range(len(string)): #loop on all nucleotide
    for l_strand in range(4, 13): # loop on the possible lenght of palindrome
        if l_strand + idx > len(string):
            continue
        rev_string=''
        for nucleo in string[idx:idx+l_strand]:
            rev_string += dico_reverse[nucleo]
            complementary_strand=rev_string[::-1]
            if string[idx:idx+l_strand] == complementary_strand:
                liste_palindrome.append([idx+1, l_strand])
for palindrome in liste_palindrome:
    print(str(palindrome[0])+' ' +str(palindrome[1]))
```
