```python
k = 15 # dominant homoz
m = 28 # heteroz
n = 26 # recessive homoz

T = k + m + n
# Calcul of the proba to not be dominant phenotype:
Proba_res = (n/T)*((n-1)/(T-1)+m/((T-1)*2))
Proba_het = 0.5*(m/T)*(n/(T-1)+(m-1)/((T-1)*2))
# Print proba for being a dominant phenotype
print(1-(Proba_res+Proba_het))
```

```python
def open_fasta(path, mode):
    fasta_file = open(path, mode)
    sequences = {}

    for line in fasta_file:
        line = line.strip()  # Remove whitespace at the beginning and end of the line
        if line.startswith(">"):  # Check if the line is a description line
            current_sequence_name = line[1:]  # Remove the ">" symbol
            sequences[current_sequence_name] = ""  # Add the sequence name to the dictionary
        else:
            sequences[current_sequence_name] += line  # Add the sequence data to the current sequence name key in the dictionary

    fasta_file.close()
    return sequences

seqs= open_fasta(r"C:\Users\sylv_\Desktop\Rosalind download\rosalind_cons.txt", "r")
print(seqs)
```

### Mortal Fibonacci Rabbits Problem

```python
def MortalRabbits(n,m,k):
    # n is the number of month, m is the time where rabbits die, k is the number of offspring pair
    list_alive= [1,1] # List of alive rabits initialize to [1,1] for the first month(as it take 1 month to come in age of reproduction)
    for i in range (2,n):
        # Calcul of rabbits pair after reproduction phase
        n_rabbits = list_alive[i-1] + list_alive[i-2]*k
        # Remote of the deaths
        if i == m:
            n_rabbits -= 1
        if i > m:
            n_rabbits -= list_alive[i-m-1]
        list_alive.append(n_rabbits)
    return list_alive[-1]
print (MortalRabbits(94,20,1))    
            
```

```python
def fib(n,m=1):
    ages = [1] + [0]*(m-1)
    for i in range(n-1):
        ages = [sum(ages[1:])] + ages[:-1]
    return sum(ages)
```
