#### Definition and dictionaries that is used in many problems

```python
def open_fasta(path, mode):
    fasta_file = open(path, mode)
    sequences = {}

    for line in fasta_file:
        line = line.strip()  # Remove whitespace at the beginning and end of the line
        if line.startswith(">"):  # Check if the line is a description line
            current_sequence_name = line[1:]  # Remove the ">" symbol
            sequences[current_sequence_name] = ""  # Add the sequence name to the dictionary
        else:
            sequences[current_sequence_name] += line  # Add the sequence data to the current sequence name key in the dictionary

    fasta_file.close()
    return sequences

def DNA_to_RNA(strand):
    RNA=''
    for char in (strand):
        if char == 'T':
            RNA += 'U'
        else:
            RNA += char
    return RNA

def Traduction_to_protein(RNA_chunk):
    dico_RNA_to_prot={'UUU' :'F','CUU' : 'L', 'AUU' : 'I', 'GUU' : 'V',
    'UUC' : 'F', 'CUC' : 'L', 'AUC' : 'I', 'GUC' : 'V',
    'UUA' : 'L', 'CUA' : 'L', 'AUA' : 'I', 'GUA' : 'V',
    'UUG' : 'L', 'CUG' : 'L', 'AUG' : 'M', 'GUG' : 'V',
    'UCU' : 'S', 'CCU' : 'P', 'ACU' : 'T', 'GCU' : 'A',
    'UCC' : 'S', 'CCC' : 'P', 'ACC' : 'T', 'GCC' : 'A',
    'UCA' : 'S', 'CCA' : 'P', 'ACA' : 'T', 'GCA' : 'A',
    'UCG' : 'S', 'CCG' : 'P', 'ACG' : 'T', 'GCG' : 'A',
    'UAU' : 'Y', 'CAU' : 'H', 'AAU' : 'N', 'GAU' : 'D',
    'UAC' : 'Y', 'CAC' : 'H', 'AAC' : 'N', 'GAC' : 'D',
    'UAA' : ' ', 'CAA' : 'Q', 'AAA' : 'K', 'GAA' : 'E',
    'UAG' : ' ', 'CAG' : 'Q', 'AAG' : 'K', 'GAG' : 'E',
    'UGU' : 'C', 'CGU' : 'R', 'AGU' : 'S', 'GGU' : 'G',
    'UGC' : 'C', 'CGC' : 'R', 'AGC' : 'S', 'GGC' : 'G',
    'UGA' : ' ', 'CGA' : 'R', 'AGA' : 'R', 'GGA' : 'G',
    'UGG' : 'W', 'CGG' : 'R', 'AGG' : 'R', 'GGG' : 'G',}
    n = 3 
    proteins=''
    codons = [RNA_chunk[i:i+n] for i in range(0, len(RNA_chunk),n)]
    for codon in codons:
        proteins += dico_RNA_to_prot[codon]
    return proteins
```

### Enumerating k-mers Lexicographically Problem

```python
from itertools import product
# Generate the combinations of lenght n of a defined list

liste_order= ['A','B','C', 'D', 'E','F']
n=3
combinations = product(liste_order, repeat=n)
    
# Concatenate the combinations into strings and sort them lexicographically
strings = sorted(''.join(combo) for combo in combinations)
for string in strings:
    print(string)
```

##### Easy-way solution (go straight to the point with the "overlap by more than half their length" statement so just one suite to a given strand will be possible)

```python
def get_superstring(reads_list, superstring=''):
    if len(reads_list) == 0:
        return superstring

    elif len(superstring) == 0:
        superstring = reads_list.pop(0)
        return get_superstring(reads_list, superstring)

    else:
        for current_read_index in range(len(reads_list)):
            current_read = reads_list[current_read_index]
            current_read_length = len(current_read)

            for trial in range(current_read_length // 2):
                overlap_length = current_read_length - trial

                if superstring.startswith(current_read[trial:]):
                    reads_list.pop(current_read_index)
                    return get_superstring(reads_list, current_read[:trial] + superstring)

                if superstring.endswith(current_read[:overlap_length]):
                    reads_list.pop(current_read_index)
                    return get_superstring(reads_list, superstring + current_read[overlap_length:])


if __name__ == "__main__":

    small_dataset = """>Rosalind_56
ATTAGACCTG
>Rosalind_57
CCTGCCGGAA
>Rosalind_58
AGACCTGCCG
>Rosalind_59
GCCGGAATAC
"""
    substrings=open_fasta(r"C:\Users\sylv_\Desktop\Rosalind download\rosalind_long.txt","r")
    reads=[x for x in substrings.values()]
    ov = get_superstring(reads)
    print(ov)
```

### Perfect Matchings and RNA Secondary Structures Problem

```python
import numpy as np
def perfect_match(string):
    l=len(string)
    CG_count=0
    AU_count=0
    for char in string:
        if char == 'C':
            CG_count += 1
        if char == 'A':
            AU_count += 1
    perf_match= np.math.factorial(AU_count)* np.math.factorial(CG_count)
    return perf_match
```

### Transitions and Transversions Problem

```python
seqs= open_fasta(r"C:\Users\sylv_\Desktop\Rosalind download\rosalind_tran.txt", "r")

strand_1=list(seqs.values())[0]
strand_2=list(seqs.values())[1]

dico_t={('A','G') : 'transition', ('C','T'): 'transition', ('A','T'):'transversions',('G','T') : 'transversions', ('A','C') : 'transversions' , ('C','G') : 'transversions'}
transition = 0
transversion = 0
for char_1,char_2 in zip(strand_1,strand_2):
    if char_1 != char_2:
        if dico_t[tuple(sorted([char_1, char_2]))] == 'transition':
            transition +=1
        if dico_t[tuple(sorted([char_1, char_2]))] == 'transversions':
            transversion +=1
print(transition)
print(transversion)
print(transition/transversion)
```

### Completing a Tree Problem

```python
def adj_matrice(path, mode):
    
    fasta_file = open(path, mode)
    edges=''
    for idx,line in enumerate(fasta_file):
        line = line.strip()
        if idx==0:
            n = line
        if idx!=0:      
            edges += "(" +str(line) + ")"
    lst = []
    for substr in edges.split(')('):
        substr = substr.replace('(', '').replace(')', '')
        pair = tuple(map(int, substr.split()))
        lst.append(pair)

    fasta_file.close()
    return n,lst
```

```python
file=adj_matrice(r"C:\Users\sylv_\Desktop\Rosalind download\rosalind_tree.txt", "r")
print(list(file)[0])
print(list(file)[1])
```

```python
import networkx as nx
adj_matrix= list(file)[1]
num_nodes = int(list(file)[0])
# create a graph with adjacency matrix adj_matrix
G = nx.Graph(adj_matrix)

# add isolated nodes to the graph
for i in range(1,num_nodes+1):
    if i not in G.nodes():
        G.add_node(i)

# get the minimum spanning tree
T = nx.minimum_spanning_tree(G)

# get the number of edges in the minimum spanning tree
num_edges = T.size()

# calculate the number of edges needed to connect all nodes
num_nodes = len(G.nodes())
num_edges_needed = num_nodes - 1 - num_edges

print("Number of edges needed to connect all nodes:", num_edges_needed)
```
